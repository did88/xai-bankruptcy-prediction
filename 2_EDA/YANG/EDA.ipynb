```python
# 📦 필수 라이브러리
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import platform

# ✅ 한글 깨짐 방지
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 📂 데이터 불러오기
df = pd.read_csv('../../1_preprocessing/YANG/preprocessed_data.csv')

# 📌 기본 정보
print("✅ 데이터 shape:", df.shape)
print("\n📋 컬럼 목록:\n", df.columns.tolist())
print("\n❓ 결측치 현황:\n", df.isnull().sum())

# ⚙️ 수치형 컬럼 추출 (거래소코드 제외)
numeric_cols = df.select_dtypes(include=np.number).columns.drop('거래소코드')

# 📊 기초 통계 요약
display(df[numeric_cols].describe().T)

# 🔗 상관관계 히트맵
plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title("🔗 수치형 변수 간 상관관계 히트맵")
plt.show()

# 📊 Annotated Bar Plot: is_defaulted 분포
count_series = df['is_defaulted'].value_counts()
percent_series = df['is_defaulted'].value_counts(normalize=True) * 100

plt.figure(figsize=(6, 4))
ax = sns.barplot(x=count_series.index, y=count_series.values)
plt.title("📉 상장폐지 여부 분포 (is_defaulted)")
plt.xlabel("is_defaulted (0 = 정상, 1 = 상장폐지)")
plt.ylabel("회사 수")
plt.xticks([0, 1], ['정상', '상장폐지'])

# 바 위에 비율 텍스트 표시
for i, val in enumerate(count_series.values):
    percent = percent_series[i]
    ax.text(i, val + 50, f'{percent:.2f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
