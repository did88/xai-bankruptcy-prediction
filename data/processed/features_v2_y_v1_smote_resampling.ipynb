```python
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

```python
# Load the data
df = pd.read_csv("features_v2_y_v1.csv")
df.head()
```

```python
# Drop unnecessary columns
df_clean = df.drop(columns=["Unnamed: 0", "corp_nm", "stock_code", "year"])

# Separate features and target
X = df_clean.drop("is_defaulted", axis=1)
y = df_clean["is_defaulted"]

# Replace inf/-inf with NaN, then drop missing values
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.dropna(inplace=True)
y = y[X.index]  # Align target with feature index
```

```python
# Apply SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Combine into one DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled["is_defaulted"] = y_resampled
```

```python
# Save the resampled data to Excel
df_resampled.to_excel("resampled_data.xlsx", index=False)
print("Resampled data saved to resampled_data.xlsx")
```
