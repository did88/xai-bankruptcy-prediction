```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import platform
```

```python
# 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')  # 리눅스 또는 Colab 환경
```

```python
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 데이터 로드
file_path = "ts20002.xlsx"
df = pd.read_excel(file_path)
```

```python
# 기본 정보 출력
print("데이터프레임 정보:")
print(df.info())
```

```python
# 통계 요약
print("\n기초 통계 요약:")
print(df.describe(include='all'))
```

```python
# 결측치 확인
print("\n결측치 개수:")
print(df.isnull().sum())
```

```python
# 범주형 변수 분포
print("\n범주형 변수 분포:")
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\n{col} 분포:")
    print(df[col].value_counts())
```

```python
# 수치형 변수 상관관계 및 히트맵
numeric_df = df.select_dtypes(include='number')
print("\n수치형 변수 간 상관관계:")
correlation_matrix = numeric_df.corr(method='pearson')
print(correlation_matrix)
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title("상관 행렬 히트맵 (Pearson)")
plt.show()
```

```python
# 히스토그램
print("\n수치형 변수 히스토그램:")
for col in numeric_df.columns:
    plt.figure()
    sns.histplot(numeric_df[col].dropna(), kde=True)
    plt.title(f"{col} 분포")
    plt.xlabel(col)
    plt.ylabel("빈도수")
    plt.show()
```

```python
# 박스 플롯 & 바이올린 플롯
print("\n수치형 변수 박스플롯 및 바이올린 플롯:")
for col in numeric_df.columns:
    plt.figure()
    sns.boxplot(y=numeric_df[col])
    plt.title(f"{col} 박스플롯")
    plt.show()
    plt.figure()
    sns.violinplot(y=numeric_df[col])
    plt.title(f"{col} 바이올린 플롯")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# 수치형 데이터 선택 (예: df가 전체 데이터프레임일 때)
numeric_df = df.select_dtypes(include='number')

# Q-Q 플롯 시각화
print("\nQ-Q 플롯 (정규성 확인):")
for col in numeric_df.columns:
    plt.figure()
    stats.probplot(numeric_df[col].dropna(), dist="norm", plot=plt)
    plt.title(f"{col} Q-Q 플롯")
    plt.show()
```
