```python
import pandas as pd
# 1. 데이터 불러오기
raw_path = '../../0_data/raw/TS2000_RAW.xlsx'
ts = pd.read_excel(raw_path, dtype={'거래소코드':str})
ts.head()
```

```python
# target
# 2024년 상장폐지나, NaN은 0
# 아니면 1

from datetime import datetime

def is_before_2024(x):
    # x가 None 또는 NaN이면 False 반환
    if pd.isnull(x):
        return 0
    # 문자열을 datetime으로 변환 (자동 포맷 인식)
    dt = pd.to_datetime(x)
    # 기준일과 비교
    return 1 if dt < datetime(2024, 1, 1) else 0

ts2 = ts.copy()
ts2['is_defaulted'] = ts2['상장폐지일'].apply(is_before_2024)
ts2.drop(columns='상장폐지일', inplace=True)
ts2.head()
```

```python
def mark_last_defaulted(group):
    # is_defaulted 1인 연도가 있으면, 그 중 가장 마지막 연도만 1로
    if group['is_defaulted'].sum() > 0:
        max_year = group.loc[group['is_defaulted'] == 1, '회계년도'].max()
        # 모든 target을 0으로 초기화
        group['is_defaulted'] = 0
        # 마지막 연도만 1로 설정
        group.loc[group['회계년도'] == max_year, 'is_defaulted'] = 1
    return group

ts2 = ts2.groupby('거래소코드', group_keys=False).apply(mark_last_defaulted)

# 검증: is_defaulted 1이면서 해당 기업의 마지막 연도가 맞는지 확인
check = ts2[ts2['is_defaulted'] == 1].copy()
check['is_last_year'] = check.groupby('거래소코드')['회계년도'].transform('max') == check['회계년도']
print(check[~check['is_last_year']])  # 이 결과가 비어 있으면 정상입니다.
```

```python
import pandas as pd
import numpy as np
from scipy.stats.mstats import winsorize

# 1. 데이터 불러오기
# raw_path = '../../0_data/raw/TS2000_RAW.xlsx'
# df = pd.read_excel(raw_path)

df = ts2.copy()
df['회계년도'] = df['회계년도'].str[:4]

df.rename(columns={
    '회사명':'corp_nm',
    '거래소코드':'stock_code',
    '회계년도':'year',
    '자산(*)(IFRS)(천원)': '자산총계',
    '비유동자산(*)(IFRS)(천원)': '비유동자산',
    '유동자산(*)(IFRS)(천원)': '유동자산',
    '자본(*)(IFRS)(천원)': '자본총계',
    '자본금(*)(IFRS)(천원)': '자본금',
    '이익잉여금(결손금)(*)(IFRS)(천원)': '이익잉여금',
    '부채(*)(IFRS)(천원)': '부채총계',
    '비유동부채 (*)(IFRS)(천원)': '비유동부채',
    '유동부채(*)(IFRS)(천원)': '유동부채',
    '현금및현금성자산(*)(IFRS)(천원)': '현금및현금성자산',
    '영업활동으로 인한 현금흐름(간접법)(*)(IFRS)(천원)': '영업활동현금흐름',
    '투자활동으로 인한 현금흐름(*)(IFRS)(천원)': '투자활동현금흐름',
    '재무활동으로 인한 현금흐름(*)(IFRS)(천원)': '재무활동현금흐름',
    '기초 현금및현금성자산(IFRS)(천원)': '기초현금및현금성자산',
    '기말 현금및현금성자산(IFRS)(천원)': '기말현금및현금성자산',
    '매출액(수익)(*)(IFRS)(천원)': '매출액',
    '매출총이익(손실)(IFRS)(천원)': '매출총이익',
    '판매비와 관리비(물류원가 등 포함)(*)(IFRS)(천원)': '판매관리비',
    '* (정상)영업손익(보고서기재)(IFRS)(천원)': '영업이익',
    '매출원가(*)(IFRS)(천원)': '매출원가',
    '당기순이익(손실)(IFRS)(천원)': '당기순이익',
    '사채(*)(IFRS)(천원)': '사채',
    '단기차입금(*)(IFRS)(천원)': '단기차입금',
    '장기차입금(IFRS)(천원)': '장기차입금',
    '재고자산(*)(IFRS)(천원)': '재고자산',
    '매출채권(IFRS)(천원)': '매출채권',
    '매출채권 및 기타유동채권(*)(IFRS)(천원)': '매출채권및기타유동채권'
}, inplace=True)
df.head()
```

```python
# 2. 행 기준 결측치 80% 이상이면 해당 행 제거
df = df[df.isnull().mean(axis=1) < 0.8]

# 3. 결측치 중앙값으로 대체 (수치형 컬럼만)
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df[col].isna().sum() > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)

# 4. 윈저라이징 (상하위 0.5%)
for col in numeric_cols:
    mask = df[col].notna()
    if mask.sum() > 0:
        df.loc[mask, col] = winsorize(df.loc[mask, col], limits=[0.005, 0.005])


# 5. 저장
output_path = 'preprocessed_data.csv'
df.to_csv(output_path, index=False)
print(f"✅ 전처리 완료: {output_path}")
```
