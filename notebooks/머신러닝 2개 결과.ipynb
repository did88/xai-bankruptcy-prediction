```python
# 한글 폰트 설정 코드 (더 확실한 방법)

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform
import warnings
warnings.filterwarnings('ignore')

print("현재 운영체제:", platform.system())

# 시스템에 설치된 모든 폰트 확인
all_fonts = [f.name for f in fm.fontManager.ttflist]
print(f"총 설치된 폰트 개수: {len(all_fonts)}")

# 한글 폰트 찾기
korean_fonts = []
korean_keywords = ['Gothic', 'Malgun', 'Nanum', 'Gulim', 'Dotum', 'Batang', 'Korean', '맑은', '나눔', '굴림', '돋움', '바탕']

for font in all_fonts:
    for keyword in korean_keywords:
        if keyword.lower() in font.lower():
            korean_fonts.append(font)
            break

# 중복 제거
korean_fonts = list(set(korean_fonts))
print("발견된 한글 폰트:", korean_fonts[:10])  # 처음 10개만 출력

# 폰트 설정 시도
font_set = False

# 방법 1: 직접 폰트 경로 지정 (Windows)
if platform.system() == 'Windows':
    import os
    windows_fonts = [
        'C:/Windows/Fonts/malgun.ttf',      # 맑은 고딕
        'C:/Windows/Fonts/gulim.ttc',       # 굴림
        'C:/Windows/Fonts/batang.ttc',      # 바탕
        'C:/Windows/Fonts/NanumGothic.ttf'  # 나눔고딕 (설치된 경우)
    ]
    
    for font_path in windows_fonts:
        if os.path.exists(font_path):
            try:
                font_prop = fm.FontProperties(fname=font_path)
                plt.rcParams['font.family'] = font_prop.get_name()
                print(f"폰트 설정 성공: {font_prop.get_name()} ({font_path})")
                font_set = True
                break
            except:
                continue

# 방법 2: 폰트명으로 직접 설정
if not font_set and korean_fonts:
    for font_name in korean_fonts:
        try:
            plt.rcParams['font.family'] = font_name
            # 테스트 플롯으로 확인
            fig, ax = plt.subplots(figsize=(1, 1))
            ax.text(0.5, 0.5, '한글테스트', fontsize=10)
            plt.close(fig)
            print(f"폰트 설정 성공: {font_name}")
            font_set = True
            break
        except:
            continue

# 방법 3: rcParams 직접 설정
if not font_set:
    try:
        # Windows 기본 한글 폰트들 시도
        font_candidates = ['Malgun Gothic', 'Microsoft YaHei', 'SimHei', 'NanumGothic', 'Gulim']
        
        for font in font_candidates:
            plt.rcParams['font.family'] = font
            try:
                # 테스트
                fig, ax = plt.subplots(figsize=(1, 1))
                ax.text(0.5, 0.5, '한글', fontsize=10)
                plt.close(fig)
                print(f"폰트 설정 성공: {font}")
                font_set = True
                break
            except:
                continue
    except:
        pass

# 마이너스 기호 설정
plt.rcParams['axes.unicode_minus'] = False

# 추가 설정
plt.rcParams['font.size'] = 10
plt.rcParams['figure.titlesize'] = 12
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 10

if font_set:
    print("✅ 한글 폰트 설정 완료!")
else:
    print("⚠️ 한글 폰트 설정 실패. 영문으로 표시됩니다.")

# 현재 폰트 확인
print(f"현재 설정된 폰트: {plt.rcParams['font.family']}")

# 폰트 테스트
try:
    fig, ax = plt.subplots(figsize=(6, 2))
    ax.text(0.1, 0.7, '한글 테스트: 자산총계, 매출액, 영업이익', fontsize=14)
    ax.text(0.1, 0.5, 'English Test: Assets, Revenue, Operating Profit', fontsize=12)
    ax.text(0.1, 0.3, '숫자 테스트: -123,456원', fontsize=12)
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')
    plt.title('폰트 테스트')
    plt.tight_layout()
    plt.show()
    print("📊 폰트 테스트 완료!")
except Exception as e:
    print(f"폰트 테스트 중 오류: {e}")

print("\n" + "="*50)
print("이제 시각화 코드를 실행하세요!")
print("="*50)
```

```python
# Financial Default Prediction Model
# 재무 데이터를 이용한 부도 예측 모델

# 필요한 라이브러리 import
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import os
import tempfile

# 한글 경로 문제 해결을 위한 환경변수 설정
os.environ['TMPDIR'] = 'C:/temp_joblib'
os.environ['TMP'] = 'C:/temp_joblib'  
os.environ['TEMP'] = 'C:/temp_joblib'

# 임시 폴더 생성
temp_folder = 'C:/temp_joblib'
if not os.path.exists(temp_folder):
    os.makedirs(temp_folder)
    print(f"임시 폴더 생성: {temp_folder}")
else:
    print(f"임시 폴더 확인: {temp_folder}")

# 머신러닝 라이브러리
from sklearn.model_selection import TimeSeriesSplit, train_test_split, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.utils import parallel_backend
import lightgbm as lgb
from imblearn.over_sampling import SMOTE

# 시각화 라이브러리
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (선택사항)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

print("라이브러리 import 완료")

# 1. CSV 파일 로드 및 전처리
print("1. 데이터 로드 및 전처리 시작...")

# CSV 파일 로드
df = pd.read_csv('ts2000_processed_missing_data.csv')
print(f"원본 데이터 크기: {df.shape}")

# 결측치 및 무한값을 0으로 처리
df = df.replace([np.inf, -np.inf], np.nan)
df = df.fillna(0)
print("결측치 및 무한값 처리 완료")

# 2. 피처 선택
print("2. 피처 선택...")

selected_features = [
    '자산총계_당기', '비유동자산_당기', '유동부채_당기', '자본총계_당기',
    '매출액_당기', '총현금흐름_당기', '영업이익_당기', '당기순이익_당기',
    '매출채권_당기', '재고자산_당기'
]

# 선택한 피처와 타깃 변수만 추출
X = df[selected_features].copy()
y = df['is_defaulted'].copy()

print(f"선택된 피처 개수: {len(selected_features)}")
print(f"타깃 변수 분포:\n{y.value_counts()}")

# 3. 시계열 분할 (TimeSeriesSplit)
print("3. 시계열 분할 시작...")

# year 컬럼을 이용한 시간 순서 정렬
df_sorted = df.sort_values('year').reset_index(drop=True)
X_sorted = df_sorted[selected_features].copy()
y_sorted = df_sorted['is_defaulted'].copy()

# TimeSeriesSplit 설정 (5-fold, 마지막 fold만 사용)
tscv = TimeSeriesSplit(n_splits=5)
splits = list(tscv.split(X_sorted))
train_idx, test_idx = splits[-1]  # 마지막 fold 사용

# 훈련/테스트 데이터 분리
X_train_ts = X_sorted.iloc[train_idx]
X_test_ts = X_sorted.iloc[test_idx]
y_train_ts = y_sorted.iloc[train_idx]
y_test_ts = y_sorted.iloc[test_idx]

print(f"TimeSeriesSplit 결과:")
print(f"훈련 데이터 크기: {X_train_ts.shape}")
print(f"테스트 데이터 크기: {X_test_ts.shape}")

# 추가로 8:2 비율로 분할
X_train, X_test, y_train, y_test = train_test_split(
    X_train_ts, y_train_ts, test_size=0.2, stratify=y_train_ts, random_state=42
)

print(f"최종 훈련 데이터 크기: {X_train.shape}")
print(f"최종 테스트 데이터 크기: {X_test.shape}")

# 4. 데이터 정규화 및 SMOTE 적용
print("4. 데이터 정규화 및 SMOTE 적용...")

# StandardScaler로 정규화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("데이터 정규화 완료")

# SMOTE 적용
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)

print(f"SMOTE 적용 후 타깃 분포:")
print(f"클래스 0: {np.sum(y_train_smote == 0)}")
print(f"클래스 1: {np.sum(y_train_smote == 1)}")

# 5. 하이퍼파라미터 튜닝
print("5. 하이퍼파라미터 튜닝 시작...")

# RandomForest 하이퍼파라미터 그리드
rf_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

# LGBM 하이퍼파라미터 그리드
lgb_param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'learning_rate': [0.01, 0.1, 0.2],
    'num_leaves': [31, 50, 100],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}

# RandomForest 튜닝 (threading 백엔드 사용)
print("RandomForest 하이퍼파라미터 튜닝...")
rf_model = RandomForestClassifier(random_state=42)

# threading 백엔드로 병렬 처리 (한글 경로 문제 해결)
with parallel_backend('threading', n_jobs=-1):
    rf_random_search = RandomizedSearchCV(
        rf_model, rf_param_grid, n_iter=20, cv=3, 
        scoring='roc_auc', random_state=42, n_jobs=-1
    )
    rf_random_search.fit(X_train_smote, y_train_smote)

print(f"RandomForest 최적 파라미터: {rf_random_search.best_params_}")
print(f"RandomForest 최적 점수: {rf_random_search.best_score_:.4f}")

# LGBM 튜닝 (threading 백엔드 사용)
print("LGBM 하이퍼파라미터 튜닝...")
lgb_model = lgb.LGBMClassifier(random_state=42, verbose=-1)

# threading 백엔드로 병렬 처리 (한글 경로 문제 해결)
with parallel_backend('threading', n_jobs=-1):
    lgb_random_search = RandomizedSearchCV(
        lgb_model, lgb_param_grid, n_iter=20, cv=3,
        scoring='roc_auc', random_state=42, n_jobs=-1
    )
    lgb_random_search.fit(X_train_smote, y_train_smote)

print(f"LGBM 최적 파라미터: {lgb_random_search.best_params_}")
print(f"LGBM 최적 점수: {lgb_random_search.best_score_:.4f}")

# 6. 두 모델 모두 평가
print("6. 두 모델 모두 평가...")

# RandomForest 예측
rf_best_model = rf_random_search.best_estimator_
rf_y_pred = rf_best_model.predict(X_test_scaled)
rf_y_pred_proba = rf_best_model.predict_proba(X_test_scaled)[:, 1]

# LGBM 예측
lgb_best_model = lgb_random_search.best_estimator_
lgb_y_pred = lgb_best_model.predict(X_test_scaled)
lgb_y_pred_proba = lgb_best_model.predict_proba(X_test_scaled)[:, 1]

# 7. 모델 평가 결과
print("7. 모델 평가 결과")
print("="*70)

# RandomForest 결과
print("\n🌲 RandomForest 결과:")
print("-" * 40)
rf_cm = confusion_matrix(y_test, rf_y_pred)
rf_roc_auc = roc_auc_score(y_test, rf_y_pred_proba)

print("Confusion Matrix:")
print(rf_cm)
print("\nClassification Report:")
print(classification_report(y_test, rf_y_pred))
print(f"ROC AUC Score: {rf_roc_auc:.4f}")

# LGBM 결과
print("\n🚀 LGBM 결과:")
print("-" * 40)
lgb_cm = confusion_matrix(y_test, lgb_y_pred)
lgb_roc_auc = roc_auc_score(y_test, lgb_y_pred_proba)

print("Confusion Matrix:")
print(lgb_cm)
print("\nClassification Report:")
print(classification_report(y_test, lgb_y_pred))
print(f"ROC AUC Score: {lgb_roc_auc:.4f}")

# 모델 성능 비교 요약
print("\n📊 모델 성능 비교:")
print("-" * 40)
print(f"RandomForest - CV Score: {rf_random_search.best_score_:.4f}, Test ROC AUC: {rf_roc_auc:.4f}")
print(f"LGBM         - CV Score: {lgb_random_search.best_score_:.4f}, Test ROC AUC: {lgb_roc_auc:.4f}")

# 더 좋은 모델 선택
if rf_roc_auc > lgb_roc_auc:
    best_model_name = "RandomForest"
    best_roc_auc = rf_roc_auc
else:
    best_model_name = "LGBM"
    best_roc_auc = lgb_roc_auc

print(f"\n🏆 최고 성능 모델: {best_model_name} (ROC AUC: {best_roc_auc:.4f})")

# 시각화 - 두 모델 비교
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# RandomForest 시각화
sns.heatmap(rf_cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Not Default', 'Default'],
            yticklabels=['Not Default', 'Default'], ax=axes[0, 0])
axes[0, 0].set_title('RandomForest - Confusion Matrix')
axes[0, 0].set_xlabel('Predicted')
axes[0, 0].set_ylabel('Actual')

# LGBM 시각화
sns.heatmap(lgb_cm, annot=True, fmt='d', cmap='Greens', 
            xticklabels=['Not Default', 'Default'],
            yticklabels=['Not Default', 'Default'], ax=axes[1, 0])
axes[1, 0].set_title('LGBM - Confusion Matrix')
axes[1, 0].set_xlabel('Predicted')
axes[1, 0].set_ylabel('Actual')

# ROC Curves 비교
rf_fpr, rf_tpr, _ = roc_curve(y_test, rf_y_pred_proba)
lgb_fpr, lgb_tpr, _ = roc_curve(y_test, lgb_y_pred_proba)

axes[0, 1].plot(rf_fpr, rf_tpr, color='blue', lw=2, 
                label=f'RandomForest (AUC = {rf_roc_auc:.4f})')
axes[0, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[0, 1].set_xlim([0.0, 1.0])
axes[0, 1].set_ylim([0.0, 1.05])
axes[0, 1].set_xlabel('False Positive Rate')
axes[0, 1].set_ylabel('True Positive Rate')
axes[0, 1].set_title('RandomForest - ROC Curve')
axes[0, 1].legend(loc="lower right")

axes[1, 1].plot(lgb_fpr, lgb_tpr, color='green', lw=2, 
                label=f'LGBM (AUC = {lgb_roc_auc:.4f})')
axes[1, 1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[1, 1].set_xlim([0.0, 1.0])
axes[1, 1].set_ylim([0.0, 1.05])
axes[1, 1].set_xlabel('False Positive Rate')
axes[1, 1].set_ylabel('True Positive Rate')
axes[1, 1].set_title('LGBM - ROC Curve')
axes[1, 1].legend(loc="lower right")

# 두 모델 ROC 비교
axes[0, 2].plot(rf_fpr, rf_tpr, color='blue', lw=2, 
                label=f'RandomForest (AUC = {rf_roc_auc:.4f})')
axes[0, 2].plot(lgb_fpr, lgb_tpr, color='green', lw=2, 
                label=f'LGBM (AUC = {lgb_roc_auc:.4f})')
axes[0, 2].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
axes[0, 2].set_xlim([0.0, 1.0])
axes[0, 2].set_ylim([0.0, 1.05])
axes[0, 2].set_xlabel('False Positive Rate')
axes[0, 2].set_ylabel('True Positive Rate')
axes[0, 2].set_title('모델 비교 - ROC Curves')
axes[0, 2].legend(loc="lower right")

# 성능 지표 비교 바 차트
metrics_comparison = pd.DataFrame({
    'Model': ['RandomForest', 'LGBM'],
    'CV_Score': [rf_random_search.best_score_, lgb_random_search.best_score_],
    'Test_ROC_AUC': [rf_roc_auc, lgb_roc_auc]
})

x = np.arange(len(metrics_comparison['Model']))
width = 0.35

axes[1, 2].bar(x - width/2, metrics_comparison['CV_Score'], width, 
               label='CV Score', alpha=0.8, color='skyblue')
axes[1, 2].bar(x + width/2, metrics_comparison['Test_ROC_AUC'], width, 
               label='Test ROC AUC', alpha=0.8, color='lightcoral')

axes[1, 2].set_xlabel('Models')
axes[1, 2].set_ylabel('Score')
axes[1, 2].set_title('성능 지표 비교')
axes[1, 2].set_xticks(x)
axes[1, 2].set_xticklabels(metrics_comparison['Model'])
axes[1, 2].legend()
axes[1, 2].set_ylim([0.75, 1.0])

# 성능 수치 표시
for i, (cv, test) in enumerate(zip(metrics_comparison['CV_Score'], metrics_comparison['Test_ROC_AUC'])):
    axes[1, 2].text(i - width/2, cv + 0.005, f'{cv:.4f}', ha='center', va='bottom')
    axes[1, 2].text(i + width/2, test + 0.005, f'{test:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 피처 중요도 비교
print("\n📈 피처 중요도:")
print("-" * 40)

# RandomForest 피처 중요도
rf_feature_importance = pd.DataFrame({
    'feature': selected_features,
    'rf_importance': rf_best_model.feature_importances_
}).sort_values('rf_importance', ascending=False)

print("\nRandomForest 피처 중요도:")
print(rf_feature_importance)

# LGBM 피처 중요도
lgb_feature_importance = pd.DataFrame({
    'feature': selected_features,
    'lgb_importance': lgb_best_model.feature_importances_
}).sort_values('lgb_importance', ascending=False)

print("\nLGBM 피처 중요도:")
print(lgb_feature_importance)

# 피처 중요도 시각화
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.barplot(data=rf_feature_importance, x='rf_importance', y='feature', ax=axes[0])
axes[0].set_title('RandomForest - Feature Importance')
axes[0].set_xlabel('Importance')

sns.barplot(data=lgb_feature_importance, x='lgb_importance', y='feature', ax=axes[1])
axes[1].set_title('LGBM - Feature Importance')
axes[1].set_xlabel('Importance')

plt.tight_layout()
plt.show()

print("\n모든 작업이 완료되었습니다!")
```
