## 1. 데이터 불러오기 및 기본 정보

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('연결재무제표_손익계산서.csv')
df.head()
```

## 2. 데이터 기본 정보 확인

```python
df.info()
df.describe(include='all')
```

## 3. 결측치 분석

```python
import matplotlib.pyplot as plt
import platform

# OS별 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':  # macOS
    plt.rcParams['font.family'] = 'AppleGothic'
else:  # Linux
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False  # 마이너스 부호 깨짐 방지

# 결측치 시각화
missing = df.isnull().sum().sort_values(ascending=False)
missing = missing[missing > 0]
missing.plot(kind='barh', figsize=(10, 6), title='결측치 개수')
plt.show()
```

## 4. 범주형 변수 분포

```python
categorical_cols = ['회사명', '종목코드', '사업연도']
for col in categorical_cols:
    display(df[col].value_counts().head(10))
```

## 5. 수치형 변수 히스토그램

```python
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df[numerical_cols].hist(figsize=(15, 12), bins=30)
plt.tight_layout()
plt.show()
```

## 6. 수치형 변수 박스플롯

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df[numerical_cols])
plt.xticks(rotation=90)
plt.title('박스플롯으로 분포 확인')
plt.show()
```

## 7. 상관관계 분석

```python
corr = df[numerical_cols].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('수치형 변수 상관관계')
plt.show()
```

## 8. 사업연도별 주요 지표 추이

```python
df['사업연도'] = df['사업연도'].astype(str)
grouped = df.groupby('사업연도')[['매출액', '영업이익', '당기순이익']].mean()
grouped.plot(marker='o', figsize=(12, 6), title='사업연도별 평균 재무 지표')
plt.grid(True)
plt.show()
```

## 9. 회사별 주요 재무 지표 비교

```python
top_companies = df['회사명'].value_counts().head(5).index
df_top = df[df['회사명'].isin(top_companies)]
pivot = df_top.pivot_table(index='사업연도', columns='회사명', values='영업이익')
pivot.plot(figsize=(12, 6), marker='o', title='회사별 영업이익 추이')
plt.grid(True)
plt.show()
```
