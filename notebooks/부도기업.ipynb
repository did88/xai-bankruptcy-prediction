```python
import pandas as pd

df_all = pd.read_csv("../data/raw/재무제표_재무상태표.csv")
df_all.drop(['Unnamed: 0'], axis=1, inplace=True)
df_all.head()
```

```python
company_codes = []
for code in df_all['회사코드'].astype(str).unique():
    if len(code) == 6:
        company_codes.append('00' + code)
    elif len(code) == 7:
        company_codes.append('0' + code)
    else:
        company_codes.append(code)
company_codes
```
